## What is a Digital Image?
**Grayscale image:** 2D array, values = pixel intensity(0–255 8-bit, 0-4095 12-bit, 0-65535 16-bit)

**Color image:** 3D array (height, width, channels)

### Libraries:
- `numpy` – numerical processing
- `matplotlib` – visualization
- `skimage` – image processing

**Code Snippet:** Load and display an image

```python
from skimage import io
import matplotlib.pyplot as plt

image = io.imread('data/nuclei.tif')
plt.imshow(image, cmap='gray')
plt.title("Original Image")
plt.axis('off')
plt.show()
```

---
## Basic Operations on Images
### Viewing Properties

```python
print(f"Image shape: {image.shape}")
print(f"Data type: {image.dtype}")
```

## Cropping and Slicing

```python
cropped = image[100:300, 100:300]
plt.imshow(cropped, cmap='gray')
```

### Image Inversion

```python
inverted = 255 - image
plt.imshow(inverted, cmap='gray')
```

### Rescaling Intensity

```python
from skimage.exposure import rescale_intensity

rescaled = rescale_intensity(image, in_range=(50, 200), out_range=(0, 255))
plt.imshow(rescaled, cmap='gray')
```

---
## Image Filtering and Enhancement
### Smoothing with Gaussian Filter

```python
from skimage.filters import gaussian

smooth = gaussian(image, sigma=2)
plt.imshow(smooth, cmap='gray')
```

### Edge Detection (Sobel Filter)

```python
from skimage.filters import sobel

edges = sobel(image)
plt.imshow(edges, cmap='gray')
```
