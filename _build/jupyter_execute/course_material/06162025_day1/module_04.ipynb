## Functions
**What are Functions?**

A function is a reusable block of code designed to perform a specific task. Python has many built-in functions, and you can also create your own.

### Built-in Functions (commonly used in bioimage tasks):
Built-in functions are functions that come with Python. They are always available and ready to use—no need to install or import anything.

These functions save time and effort. For example, instead of manually counting items in a list, you can use `len()`. Instead of writing a loop to find the biggest value, you can use `max()`.

**Why Use Them?**

- Save time: They are fast and optimized.
- Easy to use: They simplify common tasks.
- Readable: They make your code easier to understand.

**Common Built-in Functions**

| Function  | Description                                                                |
| --------- | -------------------------------------------------------------------------- |
| `len()`   | Returns the number of elements (e.g., pixels in a row, images in a stack). |
| `range()` | Creates a sequence of numbers, useful for loops.                           |
| `max()`   | Returns the maximum value in a list or array.                              |
| `min()`   | Returns the minimum value.                                                 |
| `type()`  | Tells you the data type of a variable.                                     |


**Example:**

````python
pixels = [10, 25, 34, 50]
print(len(pixels))     # 4
print(max(pixels))     # 50
print(type(pixels))    # <class 'list'>
````

## Custom Functions

**What is a Function?**

A function is a named block of code that performs a task. You define it once and use it multiple times. Functions reduce repetition and help organize your code.

**Syntax:**

```python
def function_name(parameters):
    # code
    return result
```

**Why Use Custom Functions?**

- Avoid repetition: No need to copy-paste the same code.
- Make code cleaner: Split complex tasks into small pieces.
- Reusability: Use the same function with different values.

**Example:**

````python
# Converts microns to pixels given a resolution (e.g., 0.5 µm/pixel)
def microns_to_pixels(microns, resolution):
    return microns / resolution

print(microns_to_pixels(10, 0.5))  # Output: 20
print(microns_to_pixels(25, 0.25)) # Output: 100
````

## Libraries in Python

**What are Libraries?**

Libraries are collections of functions and tools grouped into packages. You can import them to get extra functionality.

Python comes with many standard libraries, and you can also install more.

**Why Use Libraries?**
- Powerful tools for image analysis, math, plotting, etc.
- Avoid reinventing the wheel.
- Used in almost all real-world scientific code.

### Some libraries
| Library      | What It Does                                                  |
| ------------ | ------------------------------------------------------------- |
| `numpy`      | Works with arrays, fast math operations. Good for image data. |
| `pandas`     | Handles tables and metadata (like Excel).                     |
| `matplotlib` | Makes plots and graphs.                                       |


**Example:**

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create a NumPy array (simulate pixel values)
pixels = np.array([100, 110, 95, 105, 98])
print("Mean pixel value:", np.mean(pixels))

# Create a simple table
data = pd.DataFrame({
    'Cell': ['A', 'B', 'C'],
    'Area (µm²)': [150, 200, 180]
})
print(data)

# Plotting
plt.plot([1, 2, 3], [150, 200, 180])
plt.title("Cell Areas")
plt.xlabel("Cell Index")
plt.ylabel("Area (µm²)")
plt.show()
````

## Mini Project: Simulated Line Profile & Summary Table
Create a full workflow that:
- Generates a noisy line profile (using a custom function).
- Plots the profile (using matplotlib).
- Summarizes min, max, mean (using numpy).
- Stores results in a pandas DataFrame.

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 1. Custom function to generate line profile
def generate_profile(length, noise=5):
    base = np.linspace(100, 200, length)
    noise = np.random.normal(0, noise, length)
    return base + noise

# 2. Generate profile
profile = generate_profile(100)

# 3. Plot it
plt.plot(profile)
plt.title("Simulated Line Profile")
plt.xlabel("Pixel")
plt.ylabel("Intensity")
plt.show()

# 4. Summary using numpy
summary = {
    "Min": np.min(profile),
    "Max": np.max(profile),
    "Mean": np.mean(profile),
}

# 5. Store in DataFrame
df = pd.DataFrame([summary])
print(df)
````
